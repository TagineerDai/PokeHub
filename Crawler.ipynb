### Crawl the name list and official photo

```python
from __future__ import print_function
import requests
from bs4 import BeautifulSoup
import urllib
import os
import time
```

```python
#StartURL = 'https://www.pokemon.com/us/pokedex/bulbasaur'
StartURL = 'https://www.pokemon.com/us/pokedex/luvdisc'
prefix = 'https://www.pokemon.com'
```

```python
def gContent(url):
    if prefix not in url:
        url = prefix + url
    Content = requests.get(url).text
    return Content
```

```python
def gInfo(url):
    content = gContent(url)
    soup = BeautifulSoup(content)
    #print soup.prettify()
    poke = soup('img','active')[0]
    img = poke['src']
    # name = poke['alt'] 
    # The female notion for #029 Pokemon is not in ascii
    name = url.split('/')[-1]
    index = int(img[-7:-4])
    #print img, name
    nexts = soup('a','next')[0]['href']
    return img, index, name, nexts

(imgurl, index, name, nexts) = gInfo(StartURL)
```

```python
def gImage(url):
    os.mkdir(url[-7:-4])
    urllib.urlretrieve(url, url[-7:-4]+'/'+url[-7:])
```

```python
index = 370
url = StartURL
f = open('pokelist.txt','a')
while index < 801:
    (imgurl, index, name, nexts) = gInfo(url)
    #gImage(imgurl)
    f.write(str(index) + ' ' + name + '\n')
    print(str(index) + ' ' + name)
    time.sleep(0.5)
    url = nexts
    
f.close()
```
